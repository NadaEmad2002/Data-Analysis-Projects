# Exploratory Data Analysis

```python
# count the number of houses with unique floor values.
df['floors'].value_counts().to_frame()
```

```python
df.hist(color='c',figsize=(10,10))
```

```python
plt.scatter(df['sqft_living'], df['bedrooms'])
```

```python
#determine whether houses with a waterfront view or without a waterfront view have more price outliers.
sns.boxplot(data=df,x=df['waterfront'],y=df['price'])
```

```python
#determine if the feature sqft_above is negatively or positively correlated with price.
sns.regplot(data=df ,x= df['sqft_above'], y=df['price'],color='c')
```

```python
c = df.corr()
c
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(c,cmap='YlGnBu',linewidths=0.3 ,annot= True)
plt.show()
```

```python
df.drop('price',axis=1).corrwith(df['price']).plot(kind='bar',grid=True,color='y',title="Correlation with price")
```
