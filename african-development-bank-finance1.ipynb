```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib inline
```

```python
data = pd.read_csv('/kaggle/input/african-development-bank/finance1.csv',encoding='latin-1')
```

```python
data.head()
```

```python
data.info()
```

```python
df = data.dropna()
```

```python
df.info()
```

```python
df['Starting Date']=pd.to_datetime(df['Starting Date'])
```

```python
df.info()
```

```python
df['Starting Year']=df['Starting Date'].dt.year
```

```python
df.head()
```

```python
df1 =df[ df['Starting Year']==2019]
```

```python
df1.head()
```

```python
df1= df1.drop(['Project Code','Title','Starting Date','Status', 'Source of Financing','Sector','Sovereign'],axis=1)
```

```python
df1
```

```python
df1.info()
```

```python
df1['Commitment in U.A']=df1['Commitment in U.A'].str.replace(r',','')
```

```python
df1['Commitment in U.A']=df1['Commitment in U.A'].apply(pd.to_numeric)
```
