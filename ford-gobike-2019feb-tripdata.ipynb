# Data Cleaning 

```python
data.isna().sum()
```

```python
df =data.dropna()
```

```python
df.info()
```

```python
df['duration_hour']=df['duration_sec'] / 3600
```

```python
df.head()
```

```python
df.start_time = pd.to_datetime(df.start_time)
df.end_time = pd.to_datetime(df.end_time)
```

```python
df.info()
```

```python
 df.head()
```

# Data Analysis

```python
plt.hist(df.start_time.dt.hour,color='c',label='start_time')

plt.xlabel('Hour')
plt.ylabel('Count')

plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.hist(df.start_time.dt.day_name(),color='c',width=0.5,label='start_time')

plt.xlabel('week Days')
plt.ylabel('Count')

plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df.start_time.dt.date,color='c',label='start_time')

plt.xlabel('Date')
plt.ylabel('Count')

plt.legend()
plt.show()
```

```python
grouped_birth_yead = df.groupby('member_birth_year')['duration_hour']
birth_yead_agg = grouped_birth_yead.agg(['mean','count','max'])
```
